### Множество Жулиа (Juliaset) как разновидность множества Мандельброта

Из Википедии:
Мно́жество Мандельбро́та — это множество таких точек c на комплексной плоскости, для которых рекуррентное соотношение z_{n+1} = {z_n}^2 + c при z_0 = 0 является ограниченным. То есть, это множество таких c, для которых существует такое действительное R, что неравенство |zn|<R выполняется при всех натуральных n.

<div style="text-align: left;"><img src="https://upload.wikimedia.org/wikipedia/commons/2/28/Fractal_julia.png" style="margin: 0px 0px 5px 0px; width: 200px;" title="" alt="" /></div>

```julia
# jset.jl

function juliaset(z, z0, nmax::Int64)
    for n = 1:nmax
        if abs(z) > 2 (return n-1) end
        z = z^2 + z0
    end
    return nmax
end
```

```julia
# pgmfile.jl

function create_pgmfile(img, outf::AbstractString)
  s = open(outf, "w")
  write(s, "P5\n")
  n, m = size(img)
  write(s, "$m $n 255\n")
  for i=1:n, j=1:m
    p = img[i,j]
    write(s, p % UInt8)
  end
  close(s)
end
```

```julia
#
# jmain.jl
#

# В консоли интепретатора REPL следует сперва включить в среду 2 следующих файла:
# include("jset.jl")
# include("pgmfile.jl")

cd(joinpath(homedir(), "projects","mastering_julia","images"))

h = 400; w = 800; m = Array(Int64, h, w);
c0 = -0.8+0.16im;

pgm_name = "julia.pgm";

t0 = time();

for y=1:h, x=1:w
    c = complex((x-w/2)/(w/2), (y-h/2)/(w/2))
    m[y,x] = juliaset(c, c0, 256)
end

t1 = time();
   
create_pgmfile(m, pgm_name);

print("Записан $pgm_name\nВыполнено за $(t1-t0) сек.\n");
```
