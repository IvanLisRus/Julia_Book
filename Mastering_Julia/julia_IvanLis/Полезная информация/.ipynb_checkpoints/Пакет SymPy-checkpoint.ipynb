## Пакет SymPy
### Примеры

```julia
using SymPy
x = Sym("x")
```

```julia
using Gadfly

tangent(f, c) = x -> f(c) + f'(c)*(x-c)
f(x) = x^x
c = 1
Gadfly.plot([f, tangent(f, c)], 1/2, 2)
```

```julia
using Gadfly

f(x) = 2 - x^2
c = -0.75
sec_line(h) = x -> f(c) + (f(c + h) - f(c))/h * (x - c)
Gadfly.plot([f, sec_line(1), sec_line(.75), sec_line(.5), sec_line(.25)], -1, 1)
```
