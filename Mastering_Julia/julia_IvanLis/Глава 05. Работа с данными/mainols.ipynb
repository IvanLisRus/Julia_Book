## Использование нативного модуля ols.jl

```julia
cd(joinpath(homedir(), "julia_projects","data"))
include("../code/ols.jl")

data = readcsv("gasoline.csv")

# Взять столбцы, которые соответствуют нашим значениям X
# и создать новую матрицу, которая будет их содержать
# синтаксис: data[строка, столбец]
# получить всю строку = data[строка, :]
# получить весь столбец = data[:, столбец]
# получить диапазон столбцов = data[:, столбец_a:столбец_b]

x = [data[:,2] data[:,3] data[:,4]]
y = data[:,6]

# Создать новый объект OLS для линейной аппроксимации на основе метода МНК

# оценка детонационной стойкости топлива
reg = ols.tols(y, x, "Октановая характеристика", ["Ошибка", "Компонент 1", "Компонент 2", "Компонент 3"])
```

```julia
ols.summary(reg)
```
