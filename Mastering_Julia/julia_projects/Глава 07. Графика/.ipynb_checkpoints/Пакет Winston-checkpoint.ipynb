### Пакет Winston

```julia
# определим глобальные переменные

f(x::LinSpace{Float64}) = 10x.*exp(-0.3x).*sin(3x);
g(x::LinSpace{Float64}) = 0.03x.*(2pi - x).*(4pi - x);
h(x::LinSpace{Float64}) = 1./(1 + x.^2);

t = linspace(0,4pi,1000);

y1 = f(t); 
y2 = g(t); 
y3 = h(t);
```

```julia
using Winston

plot(t,y1,"b--")
```

```julia
using Winston

t = linspace(0,4pi,1000);
fplot(x->10*x.*exp(-0.3*x).*sin(3*x), [0,4pi],"b-")
```

```julia
using Winston

a = randn(10000)
ha = hist(a,100)
plothist(ha)
```

```julia
using Winston

# plot(t, y1, "r-", t, y2, "b-")

plot(t,y1, "r-")
oplot(t,y2, "b-")
plot(t,y1, "r-")
hold()
plot(t,y2, "b-")
```

```julia
using Winston

semilogx(t,y1)
title("log(t) vs 10x * exp(-0.3x) * sin(3x)")
```

```julia
using Winston

semilogy(y2,y3)
title("0.03*(2\\pi - x)*(4\\pi - x) vs log(1 /( 1 + x*x))")
```

```julia
using Winston

p = FramedPlot( title="Частота vs Время",
                          ylabel="\\Theta",
                          xlabel="\\Tau");
add(p,FillBetween(t,y1,t,y2));
add(p,Curve(t,y1,color="red"));
add(p,Curve(t,y2,color="blue"));
display(p)
```

```julia
#
# winston_graph.jl
#

cd(joinpath(homedir(),"julia_projects","images"))

using Winston

p = FramedPlot(aspect_ratio=1,xrange=(0,100),yrange=(0,100));
n = 21;
x = linspace(0, 100, n);

# Создать множество случайных величин
yA = 40 .+ 10*randn(n);
yB = x .+ 5*randn(n);

# Установить метки и стили символов
a = Points(x, yA, kind="circle");
setattr(a,label="точки a");

b = Points(x, yB);
setattr(b,label=" точки b");
style(b, kind="filled circle");

# Начертить линию, которая "оптимально пролегает" через точки yB 
# и добавить легенду в верхнюю левую часть графика
s = Slope(1, (0,0), kind="dotted");
setattr(s, label="наклонная");

lg = Legend(.1, .9, {a,b,s});
add(p, s, a, b, lg);
display(p)

savefig(p, "figure_7_4.png")
```

### Пакет Winston (повторно)

```julia
# из раздела 'Пакет Winston (повторно)'

using Winston, Color

x = linspace(0.,4π,1000);
y = linspace(0.,4π,1000);
z1 = 100*(x.^0.5).*exp(-0.1y)';

colormap("jet", 10);
imagesc(z1)
```
