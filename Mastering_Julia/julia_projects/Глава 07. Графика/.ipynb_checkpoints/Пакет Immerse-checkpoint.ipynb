<div style="text-align: left;"><h1>Глава 7. Графика</h1></div>

### Пакет Immerse (работает на основе пакета Gadfly)

```julia
using Immerse, Distributions

X = rand(MultivariateNormal([0.0, 0.0], [1.0 0.5; 0.5 1.0]), 10000)

plot(x=X[1,:], y=X[2,:], Geom.hexbin)
```

Вывод в ячеку посредством пакета Gadfly

```julia
using Immerse

plot(x=rand(5), y=rand(5), Geom.point)  
```

Тот же график с выводом в отдельное окно посредством пакета Immerse

```julia
using Immerse

display(plot(x=rand(5), y=rand(5), Geom.point)) 
```

Пример с созданием двух окон

```julia
using Immerse

y = rand(10^4)
p = plot(x=1:length(y),y=y,Geom.line)

figure()
hfig = display(p)

p2 = plot(x=1:length(y),y=y,Geom.line)

figure()
hfig2 = display(p2)
```
