### Пакет PyPlot

```julia
cd(joinpath(homedir(),"julia_projects","images"))

using PyPlot

x = linspace(0,2pi,1000)
y = sin(3*x + 4*cos(2*x));

xkcd(false)
plot(x, y, color="red", linewidth=2.0, linestyle="--");

title("Синусоидально модулированная синусоида");

savefig("figure_7_7_sinusoid.svg");

print("График записан на диск.")
```

```julia
using PyPlot

y = linspace(0,3π,250)
surf(y, y, y .* sin(y) .* cos(y)' .* exp(-0.4y))
```

```julia
#
# xkcd_graph.jl
#

using PyPlot

xkcd()

fig = figure(figsize=(10,5))

ax = axes()

p = plot(x,sin(3x + cos(5x)))

ax[:set_xlim]([0.0,6])
annotate("Несколько\nсхематичный",xy=[0.98,.001],
arrowprops=["arrowstyle"=>"->"],xytext=[1.3,-0.3])
xticks([])
yticks([])
xlabel("ВРЕМЯ")
ylabel("ПРОИЗВОДИТЕЛЬНОСТЬ")
title("График в стиле веб-комикса xkcd в Julia")
ax[:spines]["top"][:set_color]("none")
ax[:spines]["right"][:set_color]("none")
```
