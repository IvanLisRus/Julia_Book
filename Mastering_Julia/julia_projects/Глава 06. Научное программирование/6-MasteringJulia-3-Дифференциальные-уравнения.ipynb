# Дифференциальные уравнения
## Решение обыкновенных дифференциальных уравнений

<span>Динамическая система ОДУ в виде линейной модели экосистемы из трех животных (расширенная система Лотка-Вольтерры)</span>

<pre>
x[1] = p[1]*x[1] - p[2]*x[1]*x[2]
x[2] = -p[3]*x[2] + p[4]*x[1]*x[2] - p[5]*x[2]*x[3]
x[3] = -p[6]*x[3] + p[7]*x[2]*x[3]
</pre>

```julia
function F(t,x,p)
  d1 = p[1]*x[1] - p[2]*x[1]*x[2]
  d2 = -p[3]*x[2] + p[4]*x[1]*x[2] - p[5]*x[2]*x[3]
  d3 = -p[6]*x[3] + p[7]*x[2]*x[3]
  [d1, d2, d3]
end
```

```julia
p = ones(7);                      # Выбрать все параметры как 1.0
x0 = collect([0.5, 1.0, 2.0]);    # Установить исходные условия
tspan = collect([0.0:0.1:10.0]);  # и период времени
```

```julia
using ODE

(t,x) = ODE.ode23((t,x) -> F(t,x,p), x0, tspan);

n = length(t);
y1 = zeros(n); [y1[i] = x[i][1] for i = 1:n];
y2 = zeros(n); [y2[i] = x[i][2] for i = 1:n];
y3 = zeros(n); [y3[i] = x[i][3] for i = 1:n];

using Winston
Winston.plot(t,y1,"b.",t,y2,"g-.",t,y3,"r--")
```

<span>Видоизменение модели</span>

<pre>
x' = a*x - b*x*y - h*xz
d1 = p[1]*x[1] - p[2]*x[1]*x[2] - p[8]*x[1]*x[3]
</pre>

```julia
function F2(t,x,p)
  d1 = p[1]*x[1] - p[2]*x[1]*x[2] - p[8]*x[1]*x[3]  ##!
  d2 = -p[3]*x[2] + p[4]*x[1]*x[2] - p[5]*x[2]*x[3]
  d3 = -p[6]*x[3] + p[7]*x[2]*x[3]
  [d1, d2, d3]
end

using ODE

x0 = collect([0.5, 1.0, 2.0]);    # Установить исходные условия
tspan = collect([0.0:0.1:10.0]);  # и период времени

pp = ones(8);
pp[5] = pp[8] = 0.5;   # Скорость на два вида, так чтобы p5 + p8 = 1 

(t,x) = ODE.ode23((t,x) -> F2(t,x,pp), x0, tspan);

n  = length(t);
y1 = zeros(n); [y1[i] = x[i][1] for i = 1:n];
y2 = zeros(n); [y2[i] = x[i][2] for i = 1:n];
y3 = zeros(n); [y3[i] = x[i][3] for i = 1:n];

using Winston

Winston.plot(t,y1,"b.",t,y2,"g-.",t,y3,"r--")
```

## Нелинейные обыкновенные дифференциальные уравнения

Экзотермическая реакция. Развитие химической реакции во времени по функции химической кинетики Аррениуса; запись формулы в терминах разницы температур x1 и концентрации реагента x2

<pre>
x1 =  x2^n * exp(x1) - a*x1
x2 = -b*x2^n * exp(x1)
</pre>

Решение системы

```julia
function exotherm(t, x, dx; n=1, a=1, b=1) # экзотермическая реакция
  p = x[2]^n * exp(x[1])
  dx[1] = p - a*x[1]
  dx[2] = -b*p
  dx
end
```

```julia
using Sundials

t = linspace(0.0,5.0,1001);

arr = [0.9, 1.2, 1.5, 1.8]
res = Any[]

for el in arr
  fexo(t,x,dx) = exotherm(t, x, dx, a=el, b=0.1);
  push!(res, Sundials.cvode(fexo, [0.0, 1.0], convert(Array, t)));   
end
```

```julia
using Winston

plot(t,res[1][:,1],t,res[2][:,1],t,res[3][:,1],t,res[4][:,1])
```

```julia
function exotherm2(t, x, dx; a=1)
  dx[1] = exp(x[1]) - a*x[1]
  dx
end
```

```julia
t = linspace(0.0,5.0,1001);

arr2 = [2.6, 2.8, 3.0, 3.2] 
res = Any[]

for el in arr2
  fexo2(t,x,dx) = exotherm2(t, x, dx, a=el);
  push!(res, Sundials.cvode(fexo2, [0.0, 1.0], convert(Array, t)));   
end
```

```julia
using Winston

plot(t,res[1][:,1],t,res[2][:,1],t,res[3][:,1],t,res[4][:,1])
```

## Дифференциальные уравнения в частных производных

Динамика изменения температуры в стандартном диффузионном уравнении с нелинейным тепловым источником. Нагревающий терм параметризован двумя константами p1 и p2

<pre>u[i] = u[i-1] - (1-2r)*u[i] + u[i+1] + p1*exp(p2*u[i])</pre>

```julia
function diffusion_eq(p)
  r = 0.01;
  n = 100;
  t = linspace(1,2n+1,2n+1);
  u = zeros(2n+1);
  u[n] = 10;   # Точка нагрева в центре
  w = zeros(5,2n+1);

  for k1 = 1:5
    for j = 1:15000
      v = copy(u);
      v[1] = p[3]*(v[1] - v[2]);
      v[2n+1] = p[3]*(v[2n] - v[2n+1]);
      for i = 2:2n
        u[i] = v[i] + r*(v[i-1] - 2*v[i] + v[i+1]) + p[1]*(exp(p[2]*v[i]) - 1);
      end
    end
    [w[k1,k2] = u[k2] for k2 = 1:2n+1];
  end 

  hold(false)
  plot(w[1,:])
  hold(true)
  add(plot(w[2,:]))
  add(plot(w[3,:]))
  add(plot(w[4,:]))
  add(plot(w[5,:]))  
end
```

```julia
using Winston
diffusion_eq([0.001, 0.01, 0.1]) # параметризованное решение
```

```julia
diffusion_eq([0.001, 0.001, 0.001]) # непараметризованное решение
```
