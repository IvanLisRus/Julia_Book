## Пакет RDatasets

```julia
using DataFrames, RDatasets

cd(joinpath(homedir(), "julia_projects","data")); 

# вспомогательная функция NaN -> NA

function nan2na(df)
    for c in eachcol(df)
        nans = find(x->isa(x,AbstractFloat) && isnan(x),c[2])
        df[nans,c[1]] = NA
    end
    df
end

mlmf = dataset("mlmRev","Gcsemv")
mlmf = @unix ? mlmf : nan2na(mlmf); # только для Windows
```

```julia
df = mlmf[complete_cases(mlmf[[:Written, :Course]]), :]
cor(df[:Written],df[:Course])  
```

```julia
x, y = df[:Written], df[:Course];

a, b = linreg(convert(Array, x), convert(Array, y))

xlabel("Письменный экзамен"); ylabel("Курсовая");

scatter(x, y);

plot(x,convert(Array{Float64,1},[a+b*i for i in x]), "r")
```

```julia
df = mlmf[complete_cases(mlmf[[:Written, :Course]]), :]

using PyPlot 

x = df[4] 
y = df[5] 

a, b = linreg(convert(Array,x),convert(Array,y)) # линейная регрессия

PyPlot.plot(x, y, "o") # вывести точки (x,y) 
PyPlot.plot(x, [a+b*i for i in x], "r") # начертить линию
```
